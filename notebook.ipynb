## Cleaning Data

### Cleaning `customers_df` table

#### Handle duplicate data

```python
print("Number of duplications: ", customers_df.duplicated().sum())
```

```python
customers_df.drop_duplicates(inplace=True)
```

```python
print("Number of duplications: ", customers_df.duplicated().sum())
```

#### Handle missing values

```python
customers_df.isna().sum()
```

```python
customers_df[customers_df.gender.isna()]
```

```python
customers_df.gender.value_counts()
```

```python
customers_df.fillna(value="Prefer not to say", inplace=True)
```

```python
customers_df.isna().sum()
```

#### Handle inaccurate value

```python
customers_df[customers_df.age == customers_df.age.max()]
```

```python
customers_df.replace({'age': customers_df.age.max()}, 70, inplace=True)
```

```python
customers_df[customers_df.age == customers_df.age.max()]
```

```python
customers_df.replace({'age': customers_df.age.max()}, 50, inplace=True)
```

```python
customers_df.describe()
```

### Cleaning `orders_df` table

#### Change the data type of the columns "order_date" and "delivery_date" to datetime

```python
orders_df.info()
```

```python
datetime_columns = ["order_date", "delivery_date"]
 
for column in datetime_columns:
  orders_df[column] = pd.to_datetime(orders_df[column])
```

```python
orders_df.info()
```
